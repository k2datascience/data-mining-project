```python
def make_binary(row):
    for i in row['ingredients']:
        row[i] = 1
        
    return row

def ingredient_columns(df):
    new_df = df.copy()

    # Get ingredients column and create list of all unique ingredients
    ingredients = [item for sublist in df.ingredients.ravel() for item in sublist]
    unique_ingredients = list(set(ingredients))

    # Append columns of all unique ingredients 
    new_df = pd.concat([new_df, pd.DataFrame(columns=unique_ingredients)])

    # Apply function to add 1
    new_df = new_df.apply(make_binary, axis=1)
    
#     new_df = new_df.drop('ingredients', 1)
#     new_df = new_df.drop('id', 1)
    new_df.id = new_df.id.astype(int)
    new_df.set_index('id', drop=True, inplace=True)
    new_df.drop(['ingredients'], 1, inplace=True)
    
    # Fill NA with 0
    new_df.fillna(0, inplace=True)
    
    return new_df
```

```python
ingredients_df = ingredient_columns(recipe_df)
```

```python
def clean_columns(df):
    col_names = list(df.columns.values)
    col_names = [str.lower(col) for col in col_names]
    col_names = [re.sub('[^a-z]', ' ', ingredient) for ingredient in col_names]    
    df.columns = col_names
    
    return df
```

```python
# Create a dictionary of ingredients per cuisine
def create_dicts(json):
    cuisine_dict = {}
    cuisines = []
    ingredients = []
    
    for i in range(len(json)):
        cuisine = json[i]['cuisine']
        item_ingredients = json[i]['ingredients']
        
        if cuisine not in cuisine_dict.keys():
            cuisines.append(cuisine)
            cuisine_dict[cuisine] = item_ingredients
        else: 
            temp_list = cuisine_dict[cuisine]
            temp_list.extend(item_ingredients)
            cuisine_dict[cuisine] = temp_list
                 
        ingredients.extend(item_ingredients)
    
    # Unique list of all ingredients
    ingredients = list(set(ingredients)) 
    num_ingredients = len(ingredients)
    num_cuisines = len(cuisines)
    
    return cuisine_dict, num_cuisines, num_ingredients, cuisines, ingredients

cuisine_dict, num_cuisines, num_ingredients, cuisines, ingredients = create_dicts(data)
```

```python
# Create matrix with ingredient counts per cuisine
def create_term_count(dictionary, num_cuisines, num_ingredients, cuisines, ingredients):
    term_count_matrix = np.zeros((num_cuisines, num_ingredients))
    i = 0
    
    for cuisine in cuisines:
        cuisine_ingredients = dictionary[cuisine]

        for ingredient in cuisine_ingredients:
            j = ingredients.index(ingredient) 
            term_count_matrix[i,j] += 1

        i += 1

    return term_count_matrix
    

counts_matrix = create_term_count(cuisine_dict, num_cuisines, num_ingredients, cuisines, ingredients)
counts_df = pd.DataFrame(counts_matrix, index=cuisines, columns=ingredients)
```

```python
# Create TF-IDF matrix to weight relative ingredient incidence
def create_tfidf(counts):
    counts = sparse.csr_matrix(counts)
    transformer = TfidfTransformer()

    # Normalizes vectors to mean 0 and std 1 and computes tf-idf
    tfidf = transformer.fit_transform(counts).toarray() 
    return tfidf

tfidf_matrix = create_tfidf(counts_matrix)
```

```python
# Use PCA to reduce to 2 dimensions
def perform_pca(matrix, cuisines):
    pca = PCA(n_components=2) 
    data_proj = pca.fit_transform(matrix)
    
    # Use dataframe for convenience
    proj_df = pd.DataFrame(data_proj, index=cuisines)
    proj_df.columns = ['PC1', 'PC2']
    
    return data_proj, proj_df

data_proj, proj_df = perform_pca(tfidf_matrix, cuisines)
```

## k-Means
* Use elbow method to determine k

```python
# k-means clustering
def perform_kmeans(data, k):
    km = KMeans(init='k-means++', n_clusters=k, n_init=10)
    km.fit(data)
    labels = km.predict(data)
    d_space = km.transform(data)
    # sum of squared error (sklearn doesn't document this well)
    inertia = km.inertia_
    
    return labels, d_space, inertia

# iterate over range of k values and compute error for each
# we will see in a minute that the inertia returned from kmeans
# is identical to this
def km_error(data, k_range=range(1,11)):
    sse = {}
    inertias = {}
    for k in k_range:
        labels, d_space, inertia = perform_kmeans(data, k)
        inertias[k] = inertia
        for d in d_space:
            sse.setdefault(k, 0)
            sse[k] += d[d.argmin()]**2
    return sse, inertias
```

```python
# Find Jaccard Similarity (One vs. All Cuisines in Cluster)
def calculate_jaccard_similarity(cuisines, cuisine_dict):
    i = 0 
    j = 0 

    effect_on_cluster = [0 for cuisine in cuisines]

    for primary_cuisine in cuisines:  
        primary_set = set(cuisine_dict[primary_cuisine])
        secondary_set = []
        secondary_set.extend(primary_set)
#         other_points = 0
        j = 0

        for secondary_cuisine in cuisines:
            if secondary_cuisine != primary_cuisine: 
                if labels[j] == labels[i]: 
                    secondary_set.extend(set(cuisine_dict[secondary_cuisine]))
#                     other_points += 1
            j += 1

        intersection = len(set(primary_set & set(secondary_set))) / float(len(set(primary_set.union(secondary_set))))
        effect_on_cluster[i] = intersection

        i += 1
        
    return effect_on_cluster
        
effect_on_cluster = calculate_jaccard_similarity(cuisines, cuisine_dict)
```

```python
# Plot of PC1 and PC2 components
# Bubble size is determined by Jaccard Similarity
def plot_principal_components(effect_on_cluster):
    i = 0
    figure_ratios = (15,20)
    x = []
    y = []
    color = []
    area = []

    # Colors: green, blue, orange, gray and purple
    color_palette = ['#009600','#2980b9', '#ff6300','#2c3e50', '#660033', '#aaa000']
#     color_palette = sns.color_palette('spectral', 6)

    plt.figure(1, figsize=figure_ratios)

    for data in data_proj:
        x.append(data[0]) 
        y.append(data[1])  
        color.append(color_palette[labels[i]]) 

        # Magnifying the bubble's sizes by the same unit
        area.append(effect_on_cluster[i] * 27000) 

        text(data[0], data[1], cuisines[i], size=10.6, horizontalalignment='center', fontweight = 'bold', color='w')
        i += 1

    plt.scatter(x, y, c=color, s=area, linewidths=2, edgecolor='w', alpha=0.80) 

#     plt.axis([-0.45, 0.65, -0.55, 0.55])
    plt.axes().set_aspect(0.8, 'box')

    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.axis('off')

    plt.show()
    
plot_principal_components(effect_on_cluster)
```

```python
# generate user data using random sampling from discrete distribution

def partition(n_parts, total):
    
    divisions = sorted(np.random.choice(range(1, total), n_parts-1))
    parts = [x - y for x,y in 
             zip(divisions + [total], [0] + divisions)]
    return parts

def create_records(data, n_records, n_users):

     # unique cuisine types and recipe id
    cuisines = data.cuisine.unique()
    recipes = data.id.unique()
    
    # create random record counts for each user
    # record counts sum to n_records
    record_counts = partition(n_users, n_records)
    
    records = []
    
    # random discrete distribution to introduce cuisine bias
    np.random.seed(321)
    p = np.random.dirichlet(np.ones(len(cuisines)), size=1).flatten()

    for uid,c in enumerate(record_counts):
        for i in range(c):
            # select random cuisine
            cuisine = np.random.choice(cuisines, 1, p=p)[0]

            # random sample from recipes of type cuisine
            recipe = data[data.cuisine == cuisine].sample(1).iloc[0, 1]

            # create user record
            record = [uid, cuisine, recipe]
            # append record
            records.append(record)
    
    np.random.shuffle(records)
    return records
```

### Content-based Recommender

### Cosine similarity to recommend recipes that are similar to one last eaten by a user

```python
# in case you want to compute it yourself:
# a functon to compute cosine similarity of two vectors
def cos_sim(v1, v2):
    """ Return cosine similarity of two vectors
    
        v1, v2: sparse vectors with same dim
    """
    norm = np.linalg.norm
    sim = np.dot(v1,v2) / (norm(v1)*norm(v2))
    
    return sim


# a very fast method of computing a cosine similarity matrix from all possible combinations
def cos_sim_matrix(M):
    from sklearn.metrics import pairwise_distances
    
    sims = 1 - pairwise_distances(M.values, metric='cosine')
    sim_matrix = pd.DataFrame(sims, index=M.index, columns=M.index).fillna(0)
    return sim_matrix
```

```python
# recipe matrix is the same as ingredients_df
# except it is for our sample of 200 recipes
recipe_matrix = ingredient_columns(recipe_sample)
recipe_matrix = clean_columns(recipe_matrix)
# get similarity matrix from recipe matrix
cos_scores = cos_sim_matrix(recipe_matrix.select_dtypes(exclude=['O']))
```

```python
def get_similar_recipes(recipe_id, top_n, cscores, recipe_data):
    """ Return best matches to a recipe based on cosine similarity
    
        recipe_id: numeric recipe id
        top_n: number of best matches to return
        cscores: cosine similarity matrix for recipes as dataframe with recipe id's
        recipe_data: the dataframe of all available recipes with ingredients
    """
    top_recipe_scores = cscores.drop(recipe_id, axis=1).loc[recipe_id, :].nlargest(top_n)
    similar_recipe_ids = top_recipe_scores.index.values
    similar_recipes = recipe_sample[recipe_sample.id.isin(similar_recipe_ids)]
    last_recipe = recipe_sample[recipe_sample.id == recipe_id]
    
    
    print('Your last recipe was\n')
    print(last_recipe,  '\n')
    print('Here are {} recipes similar to recipe {}\n'.format(top_n, recipe_id))
    print(similar_recipes)
    
    return top_recipe_scores, similar_recipes
```

### Association rules for a collaborative approach
* Consider all the cuisines of a single user as a "transaction"
* Recipes likely to be found in a transaction (user history) with another will be used for recommendations
* This will recommend top n most likely recipes based on a user's last dish

```python
from itertools import combinations

# create pairs
itemsets = combinations(recipe_matrix.index.values, 2)
```

```python
# create a set of user histories (transactions)
history = dict()
for uid in user_data.uid.unique():
    history[uid] = user_data[user_data.uid == uid].recipe_id.tolist()
transactions = [v for v in history.values()]
transactions[2:5]
```

```python
def support(iset, itemsets):
    """ Compute support for given itemset
    
    """
    # number of itemsets containing iset
    n = np.sum([1 for items in itemsets if iset.issubset(items)])
    supp = n / len(itemsets)  # support

    return supp


# create pairs
from itertools import combinations
pairs = list(combinations(recipe_matrix.index.values, 2))

supp_d = dict()
for r in recipe_matrix.index.values:
    supp = support(set([r]), transactions)
    supp_d[frozenset([r])] = supp
for p in pairs:
    supp = support(set(p), transactions)
    supp_d[frozenset(p)] = supp
```

```python
def assoc_rules(pairs):
    # for this we will ignore min conf and supp values
    # primarily because we are not seeking anything beyond size-2 pairings
    min_conf = 0.0
    
    # iterate pairs and compute conf and lift
    # and save rules
    rules_df = dict()
    i = 0
    from itertools import permutations
    perms = permutations(recipe_matrix.index.values, 2)
    for pair in pairs:
        for perm in permutations(pair, 2):
            a = perm[0]
            c = perm[1]
            z = set([a]).union(set([c]))
            rule1 = '{} -> {}'.format(a, c)
            try:
                supp = supp_d[frozenset(z)]
                conf = supp / supp_d[frozenset([a])]
                lift = conf / supp_d[frozenset([c])]
            except Exception:
                continue

            if conf > min_conf:
                rules_df[i] = {'antec_id': a, 
                               'consq_id': c, 
                               'supp': supp, 
                               'conf': conf, 
                               'lift': lift}
                i += 1

    n_rules = len(rules_df)
    if n_rules > 0:          
        rules_df = pd.DataFrame(rules_df).T[['antec_id','consq_id','supp','conf','lift']]
        rules_df[['antec_id','consq_id']] = rules_df[['antec_id','consq_id']].astype(int)
        rules_df.sort_values(by='antec_id', inplace=True)
        print('Rules Generated: {}'.format(n_rules))
    else:
        print('No rules generated!')

    return rules_df
```

```python
rules_df = assoc_rules(pairs)
```

In the above algorithm we accounted for permutations when generating rules, because while lift remains the same no matter the order of antecedent and consequent, confidence is not the same. So they were included just in case we want to use that in making recommendations.

```python
# generate recommendations based on association rules
def recommend_assoc(recipe_id, top_n, rules_df):
    rules = rules_df[rules_df.antec_id == recipe_id]
    rules.sort_values(by='lift', ascending=False, inplace=True)
    recs = rules.head(top_n)
    print('Other users who ate recipe {} also ate these recipes:\n'.format(recipe_id))
    print('Recommended recipe IDs:\n{}'.format(recs.consq_id.tolist()))
    return recs
```

### Collaborative approach with user ratings

#### Add ratings to the data

```python
from scipy.stats import skewnorm

# create user biases (a mean score for every user)
# sampled from a skewed dist
np.random.seed(123)
user_means = skewnorm.rvs(-3.5, loc=4.5, size=user_data.uid.nunique())
plt.hist(user_means, bins=30);
```

```python
user_means = pd.Series(user_means, index=user_data.uid.unique())
# create user ratings (also sampled from skewed dist)
def f(uid):
    # offset mean to account for skew
    y = int(skewnorm.rvs(-5, loc=user_means.loc[uid])+0.75)
    if y > 5:
        y = 5
    elif y < 1:
        y = 1
    return y

user_ratings = [f(uid) for uid in user_data.uid]
user_data['rating'] = user_ratings
```

#### Recommender

Get users who meet following criteria

* have similar ratings after adjusting for bias

These criteria are used to match users based on similar taste

```python
def recommend_rating_based(uid, ratings_matrix, recipe_sims, recipe_data):
    percentile_thresh_ = 0.9
    
    recipe_sims_vector = user_recipe_sims.loc[uid].drop(uid)
    
    # users who eat similar recipes
    thresh = recipe_sims_vector.quantile(percentile_thresh_)
    sim_users_recipe = recipe_sims_vector[recipe_sims_vector > thresh]
    
    # filter further to get the ones who have similar ratings
    # determine using Euclidean distance
    mask = [uid] + sim_users_recipe.index.tolist()
    M = ratings_matrix.loc[ratings_matrix.index.isin(mask)]
    u = M.loc[uid]
    X = M.loc[M.index != uid]
    d = np.sqrt(sum((X-u)**2, axis=1))
    
    similar_users = d.nsmallest(10).index.tolist()
#     return M, d, similar_users
    
    # recommend 3 recipes not tried by uid
    uid_ratings = ratings_matrix.iloc[uid]
    recipe_scores = M.loc[similar_users, M.columns.isin(uid_ratings.isnull().index)].mean()
    recipe_ids = recipe_scores.nlargest(3).index
    
    print("Based on recipes liked by users with similar taste, here are three recipes you haven't tried:\n")
    print(recipe_data[recipe_data.id.isin(recipe_ids)].to_string(index=False), '\n')
    print('The average scores from similar users are')
    print(recipe_scores.nlargest(3))
```

```python
recommend_rating_based(22, user_recipe_ratings, user_recipe_sims, recipe_df)
```
